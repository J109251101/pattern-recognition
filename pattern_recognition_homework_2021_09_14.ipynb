```python
import cv2
from google.colab.patches import cv2_imshow
cv2.__version__
```

```python
img = cv2.imread('input.jpg')
cv2_imshow(img)
```

```python
img
```

```python
import numpy as np
```

```python
res = []

for i in range(0, 3):
  val = np.reshape(img[:, :, i], -1)
  img_mean = np.mean(val)
  img_std = np.std(val)
  res.append(img_mean)
  res.append(img_std)

res
```

```python
from matplotlib.font_manager import FontProperties
import matplotlib.pyplot as plt

myfont = FontProperties(fname=r'/usr/local/lib/python3.7/dist-packages/matplotlib/mpl-data/fonts/ttf/taipei_sans_tc_beta.ttf')
```

```python
hist_b = cv2.calcHist([img], [0], None, [4], [0, 256])
hist_b = hist_b.reshape(-1)
```

```python
plt.bar(range(0, 4), hist_b, width=1.2)
plt.ylabel('B ', fontproperties=myfont)
plt.xlabel('bins', fontproperties=myfont)
plt.title('B ', fontproperties=myfont)
plt.show()
```

```python
hist_g = cv2.calcHist([img], [1], None, [4], [0, 256])
hist_g = hist_g.reshape(-1)
plt.bar(range(0, 4), hist_g, width=1.2)
plt.ylabel('G ', fontproperties=myfont)
plt.xlabel('bins', fontproperties=myfont)
plt.title('G ', fontproperties=myfont)
plt.show()
```

```python
hist_r = cv2.calcHist([img], [2], None, [4], [0, 256])
hist_r = hist_r.reshape(-1)
plt.bar(range(0, 4), hist_r, width=1.2)
plt.ylabel('R ', fontproperties=myfont)
plt.xlabel('bins', fontproperties=myfont)
plt.title('R ', fontproperties=myfont)
plt.show()
```
